```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
fig, ax = plt.subplots()
x = np.linspace(15,95)
for i in np.arange(50,70,3):
    y = 1/(1+np.exp(-0.3*(x-i)))
    ax.plot(x,y)
ax.set_xlabel('Temperature [°C]')
ax.set_ylabel("Fluorescence Ratio [AU]")
fig.savefig('fig/ndsf.png', dpi=200, bbox_inches='tight')
```
